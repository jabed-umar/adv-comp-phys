```python
import kwant
import matplotlib.pyplot as plt
import numpy as np
from Library.kpmpy import*
```

# Hexagonal (Graphene) Lattice

```python
def circle (pos):
    r = 250
    x, y = pos
    return x**2 + y**2 < r**2
```

```python
def lattice_honeycomb_hamiltonian (a, pot ,r):
    lat = kwant.lattice.honeycomb(a)
    sys = kwant.Builder ()
    sys[lat.shape(circle , (0, 0))] = pot
    sys[lat.neighbors ()] = 1
    sys = sys.finalized ()
    kwant.plotter.plot(sys)
    return sys.hamiltonian_submatrix ( sparse =True)
```

```python
k = lattice_honeycomb_hamiltonian(1,0,250)
r = np.shape(k)[0]
```

```python
# Assuming 'k', 'params', and 'operator' are defined elsewhere
x = kwant.kpm.SpectralDensity(k, params=None, operator=None, num_vectors=10, num_moments=200)
e, rho = x()

# Save data to a text file
data = np.column_stack((e, rho))  # Combine energy and density of states into one array
# np.savetxt('DoS_10_200.txt', data, header='Energy Density_of_States', comments='')
```
